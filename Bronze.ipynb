```sql
%python

if spark.catalog.tableExists('datamodeling.silver.silver_table'):
  last_load_date = spark.sql("select max(last_updated) from datamodeling.silver.silver_table").collect()[0][0]
  if last_load_date is None:
    last_load_date = "1900-01-01"
else:
  last_load_date = "1900-01-01"
```

```sql
%python

print(last_load_date)
```

```sql
%python
spark.sql(f"""select * from datamodeling.default.source_data
where order_date >'{last_load_date}'""").createOrReplaceTempView("bronze_source")
```

```sql
Select * from bronze_source
```

```sql
%sql
create or replace table datamodeling.bronze.bronze_table
as
SELECT * FROM bronze_source;
```
