**Creating Dimension tables**

Note: It is only 1 time setup.


```sql
DROP TABLE IF EXISTS datamodeling.gold.DimCustomers;
CREATE TABLE datamodeling.gold.DimCustomers (
    DimCustomerKey BIGINT GENERATED ALWAYS AS IDENTITY,
    customer_id INT,
    customer_email VARCHAR(100),
    customer_name VARCHAR(100),
    Customer_Name_Upper VARCHAR(100)
);
```

```sql
DROP TABLE IF EXISTS datamodeling.gold.DimProducts;
CREATE TABLE datamodeling.gold.DimProducts (
    DimProductKey BIGINT GENERATED ALWAYS AS IDENTITY,
    product_id INT,
    product_name VARCHAR(100),
    product_category VARCHAR(50));
```

```sql
DROP TABLE IF EXISTS datamodeling.gold.DimRegions;
CREATE TABLE datamodeling.gold.DimRegions (
    DimRegionKey BIGINT GENERATED ALWAYS AS IDENTITY,
    country VARCHAR(50)
);
```

```sql
DROP TABLE IF EXISTS datamodeling.gold.DimPayments;
CREATE TABLE datamodeling.gold.DimPayments (
    DimPaymentKey BIGINT GENERATED ALWAYS AS IDENTITY,
    payment_type VARCHAR(50)
);
```

**Fact table**

```sql
DROP TABLE IF EXISTS datamodeling.gold.FactSales;
CREATE TABLE datamodeling.gold.FactSales (
    FactSalesKey BIGINT GENERATED ALWAYS AS IDENTITY,
    DimCustomerKey BIGINT,
    DimProductKey BIGINT,
    DimRegionKey BIGINT,
    DimPaymentKey BIGINT,
    quantity INT,
    unit_price DECIMAL(10, 2),
    order_id INT,
    loaded_date TIMESTAMP
);
```
